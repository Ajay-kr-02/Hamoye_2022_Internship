# Q13

```python
df1=df.drop(['date','lights'],axis=1) # dropping columns
```

```python
df1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df1 = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns) #normalizing data set
```

```python
features_df1= normalised_df1.drop(columns=['Appliances'],axis=1)
Appliances_target=normalised_df1['Appliances']
```

```python
#Splitting of data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df1, Appliances_target, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)
#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

<b> Mean absolute error is 0.05 </b>

# Q14

```python
#Residual Sum of Squares
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

<b> RSS of linear model = 45.35 </b>

# Q15

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

<b> Root Mean Squared Error (RMSE) of linear model is 0.088 </b>

# Q16

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

# Q17

```python
def get_linear_weights_df(model, feat, col_name):
#this function returns the weight of every feature
    weights1 = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df1 = pd.DataFrame(weights1).reset_index()
    weights_df1.columns = ['Features', col_name]
    weights_df1[col_name].round(3)
    return weights_df1

linear_model_weights = get_linear_weights_df(linear_model, x_train, 'Linear_Model_Weight')
```

```python
linear_model_weights
```

From above output table is sorted linear weights and it is clearly visible that  <b> rv2 </b> and <b> rv1 </b> have minimum and maximum weights respectively.

```python
#Regularization using Ridge (L2)
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
predicted_values_ridge = ridge_reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse_ridge = np.sqrt(mean_squared_error(y_test, predicted_values_ridge))
round(rmse_ridge, 3)
```

<b>It has been observed that RMSE calculated after regularization with Ridge is equal to RMSE calculated by simple linear regression model. So there are NO changes.</b>

# Q19

```python
#LASSO-Least Absolute Shrinkage and  Selection Operator
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
#defining function for lasso weights
def get_lasso_weights_df(model, feat, col_name):
#this function returns the weight of every feature
    weights2 = pd.Series(model.coef_, feat.columns).sort_values()
    weights2_df = pd.DataFrame(weights2).reset_index()
    weights2_df.columns = ['Features', col_name]
    weights2_df[col_name].round(3)
    return weights2_df

lasso_weights = get_lasso_weights_df(lasso_reg, x_train, 'Lasso_weight')
```

```python
lasso_weights
```

From above output table of weights in lasso model it has been observed that number of non- zero weight features are 4.
